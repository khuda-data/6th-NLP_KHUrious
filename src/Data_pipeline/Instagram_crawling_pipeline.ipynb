```python
# jpg -> png -> folder 

# 기준 디렉토리 및 PDF 저장 경로 설정
base_dir = r"C:\Users\usert\Desktop\추가 데이터셋 정리\posts"
pdf_output_dir = r"C:\Users\usert\Desktop\추가 데이터셋 정리\posts\pdf"

# PDF 저장 디렉토리가 없으면 생성
os.makedirs(pdf_output_dir, exist_ok=True)

# 기준 디렉토리의 하위 폴더 탐색
for main_folder in os.listdir(base_dir):
    main_folder_path = os.path.join(base_dir, main_folder)
    
    # 최상위 폴더가 디렉토리인지 확인
    if os.path.isdir(main_folder_path):
        # PDF 저장용 서브 폴더 생성
        main_pdf_dir = os.path.join(pdf_output_dir, main_folder)
        os.makedirs(main_pdf_dir, exist_ok=True)  # 서브 폴더 생성
        
        for sub_folder in os.listdir(main_folder_path):
            sub_folder_path = os.path.join(main_folder_path, sub_folder)
            
            # 하위 폴더가 디렉토리인지 확인
            if os.path.isdir(sub_folder_path):
                images = []
                for file_name in sorted(os.listdir(sub_folder_path)):  # 파일 이름 기준 정렬
                    file_path = os.path.join(sub_folder_path, file_name)
                    
                    # JPG 파일 필터링
                    if file_name.lower().endswith('.jpg'):
                        img = Image.open(file_path).convert('RGB')  # RGB 변환
                        images.append(img)
                
                # 이미지가 있을 경우 PDF로 저장
                if images:
                    # PDF 이름에 폴더 경로 반영
                    output_pdf_name = f"{sub_folder}.pdf"
                    output_pdf_path = os.path.join(main_pdf_dir, output_pdf_name)
                    
                    # PDF 저장
                    images[0].save(output_pdf_path, save_all=True, append_images=images[1:])
                    print(f"PDF 저장 완료: {output_pdf_path}")
```

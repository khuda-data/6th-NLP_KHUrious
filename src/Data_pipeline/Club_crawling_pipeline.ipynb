```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# WebDriver 설정
driver = webdriver.Chrome()  # ChromeDriver의 경로가 PATH에 설정되어 있어야 함
driver.get("https://jajudy.khu.ac.kr/club")  # 기본 URL로 이동

# 페이지 로딩 대기
time.sleep(3)

# 클래스가 'item'인 요소 찾기
elements = driver.find_elements(By.CLASS_NAME, 'item')

# 결과 데이터를 담을 리스트
results = []

# 데이터 추출
for element in elements:
    url = element.get_attribute('href')  # href 추출
    if not url:
        continue  # href가 없는 경우 스킵

    # URL 접속
    driver.get(url)
    time.sleep(2)  # 페이지 로딩 대기

    # 데이터 추출
    data = {}
    data['URL'] = url
    try:
        data['Name'] = driver.find_element(By.CLASS_NAME, 'name').text
    except:
        data['Name'] = None

    try:
        data['Purpose'] = driver.find_element(By.CLASS_NAME, 'purpose').text
    except:
        data['Purpose'] = None

    try:
        data['Executive'] = driver.find_element(By.CLASS_NAME, 'executive').text
    except:
        data['Executive'] = None

    try:
        data['Section_Club'] = driver.find_element(By.CLASS_NAME, 'section_club').text
    except:
        data['Section_Club'] = None

    # 텍스트 형식으로 저장
    result_text = f"""
URL: {data['URL']}
Name: {data['Name'] or 'N/A'}
Purpose: {data['Purpose'] or 'N/A'}
Executive: {data['Executive'] or 'N/A'}
Section Club: {data['Section_Club'] or 'N/A'}
"""
    results.append(result_text.strip())

# 추출된 데이터를 txt 형식으로 출력
print("\n\n==== 최종 데이터 ====\n")
final_output = "\n\n".join(results)
print(final_output)

# 드라이버 종료
driver.quit()
```

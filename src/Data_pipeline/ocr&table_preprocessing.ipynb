```python
from PyPDF2 import PdfReader, PdfWriter
import os

def split_pdf(input_pdf_path, output_dir, pages_per_split=100):
    # PDF 읽기
    reader = PdfReader(input_pdf_path)
    total_pages = len(reader.pages)
    
    # 출력 디렉토리 생성
    os.makedirs(output_dir, exist_ok=True)
    
    # 분할 및 저장
    for start_page in range(0, total_pages, pages_per_split):
        writer = PdfWriter()
        end_page = min(start_page + pages_per_split, total_pages)
        
        for page_num in range(start_page, end_page):
            writer.add_page(reader.pages[page_num])
        
        output_pdf_path = os.path.join(output_dir, f"part_{start_page + 1}_to_{end_page}.pdf")
        with open(output_pdf_path, "wb") as output_pdf:
            writer.write(output_pdf)
        print(f"{output_pdf_path} 저장 완료.")

# 사용 예시
input_pdf_path = r"C:\Users\USER\Desktop\KHURIOUS_Project\2024_교육과정_cleaning.pdf"  # 원본 PDF 파일 경로
output_dir = r"C:\Users\USER\Desktop\KHURIOUS_Project\pdf_splitter"              # 분할된 PDF 파일을 저장할 디렉토리
pages_per_split = 100                    # 분할할 페이지 수

split_pdf(input_pdf_path, output_dir, pages_per_split)
```

```python
# json-> html
import json

def get_html(data_file, excludes=[]):
    # JSON 파일 열기
    with open(data_file, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    # HTML 내용 추출
    htmls = []
    for element in data["elements"]:
        if element["category"] not in excludes:
            page = element.get("page", "unknown")  # page 번호가 없으면 "unknown"
            print(f"------------------ page {page}")
            html_content = element.get("content", {}).get("html")  # content에서 html 키 추출
            if html_content:  # html 내용이 있으면 추가
                htmls.append(html_content)
    return "\n".join(htmls)

def get_merged_html(data_files, excludes=[]):
    merged_html = ""
    for data_file in sorted(data_files):
        print(f"Processing: {data_file}")
        html = get_html(data_file, excludes)
        merged_html += f"{html}\n"
    return merged_html

# JSON 파일 경로
json_file_path = r"C:\Users\usert\Desktop\KHURIOUS_Project\files\part_1_to_100.json"

# 병합된 HTML 생성
merged_html = get_merged_html([json_file_path], excludes=["footer", "figure"])

# 결과를 HTML 파일로 저장
output_file = r"C:\Users\usert\Desktop\NLP_project\merged_output.html"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(merged_html)
print(f"Merged HTML saved to {output_file}")
```

```python
# html -> md 
import os
from bs4 import BeautifulSoup
import markdownify

def html_to_markdown(html_content):
    """
    HTML 콘텐츠를 Markdown 형식으로 변환합니다.
    """
    # BeautifulSoup으로 HTML 파싱
    soup = BeautifulSoup(html_content, 'html.parser')

    # Markdownify를 사용하여 Markdown으로 변환
    markdown_text = markdownify.markdownify(str(soup), heading_style="ATX")
    return markdown_text

def convert_html_files_in_directory(input_directory, output_directory):
    """
    주어진 디렉토리의 HTML 파일들을 Markdown 파일로 변환하여 저장합니다.
    """
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # 입력 디렉토리에서 HTML 파일 읽기
    for filename in os.listdir(input_directory):
        if filename.endswith(".html"):  # HTML 파일만 처리
            input_path = os.path.join(input_directory, filename)
            output_path = os.path.join(output_directory, f"{os.path.splitext(filename)[0]}.md")

            # HTML 파일 읽기
            with open(input_path, "r", encoding="utf-8") as f:
                html_content = f.read()

            # HTML을 Markdown으로 변환
            markdown_text = html_to_markdown(html_content)

            # 변환된 Markdown 저장
            with open(output_path, "w", encoding="utf-8") as f:
                f.write(markdown_text)
            print(f"변환 완료: {input_path} -> {output_path}")

# 예제 실행
if __name__ == "__main__":
    # 디렉토리 경로 설정
    input_directory = r"C:\Users\usert\Desktop\KHURIOUS_Project\no_table"  # HTML 파일이 저장된 디렉토리
    output_directory = r"C:\Users\usert\Desktop\KHURIOUS_Project\no_table"  # 변환된 Markdown 파일이 저장될 디렉토리

    # HTML 파일을 변환하여 Markdown 파일로 저장
    convert_html_files_in_directory(input_directory, output_directory)
```

```python
import os
import json
from bs4 import BeautifulSoup
import markdownify


def get_html_with_page_separator(data_file, excludes=[]):
    with open(data_file, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    htmls = []
    previous_page = None  # 이전 페이지 번호 추적

    for element in data.get("elements", []):
        if element.get("category") not in excludes:
            current_page = element.get("page", "unknown")  # 현재 페이지 번호 가져오기
            if previous_page is not None and current_page != previous_page:
                htmls.append("\n---\n")  # 페이지 번호가 변경되면 --- 추가

            previous_page = current_page  # 페이지 업데이트
            html_content = element.get("content", {}).get("html")  # HTML 내용 추출
            if html_content:
                htmls.append(html_content)
    return "\n".join(htmls)

def remove_table_tags(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    for table in soup.find_all("table"):
        table.decompose()  # <table> 태그와 그 내용을 제거
    return str(soup)

def get_merged_html(data_files, excludes=[]):
    merged_html = ""
    for data_file in sorted(data_files):
        print(f"Processing file: {data_file}")
        html = get_html_with_page_separator(data_file, excludes)
        merged_html += html
    return merged_html


def html_to_markdown(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    markdown_text = markdownify.markdownify(str(soup), heading_style="ATX")
    return markdown_text


def save_merged_markdown_by_ranges(input_base_path, ranges, output_base_path, excludes=[]):
    os.makedirs(output_base_path, exist_ok=True)  # 출력 디렉토리 생성

    for start, end in ranges:
        # 범위에 맞는 JSON 파일 경로 생성
        file_name = f"part_{start}_to_{end}.json"
        data_file = os.path.join(input_base_path, file_name)
        if not os.path.exists(data_file):
            print(f"File not found: {data_file}")
            continue

        print(f"Processing range: {start}-{end}")
        # 병합된 HTML 생성
        merged_html = get_merged_html([data_file], excludes=excludes)
        # HTML -> Markdown 변환
        markdown_text = html_to_markdown(merged_html)

        # 저장할 Markdown 파일 경로
        output_file = os.path.join(output_base_path, f"merged_{start}_to_{end}.md")
        with open(output_file, "w", encoding="utf-8") as f:
            f.write(markdown_text)
        print(f"Markdown 저장 완료: {output_file}")



```

```python
import os
import re
#병합코드 
def merge_markdown_files(input_directory, output_file):
    files = [f for f in os.listdir(input_directory) if f.endswith(".md")]

    def extract_range(file_name):
        match = re.search(r"(\d+)_to_(\d+)", file_name)
        if match:
            return int(match.group(1)), int(match.group(2))
        return float('inf'), float('inf')  # 매칭되지 않는 파일은 맨 뒤로 정렬

    sorted_files = sorted(files, key=extract_range)


    with open(output_file, "w", encoding="utf-8") as outfile:
        for file_name in sorted_files:
            file_path = os.path.join(input_directory, file_name)
            with open(file_path, "r", encoding="utf-8") as infile:
                content = infile.read()
                outfile.write(content + "\n\n") 
                print(f"Merged: {file_name}")

    print(f"All files merged into {output_file}")


if __name__ == "__main__":
    input_directory = r"C:\Users\usert\Desktop\KHURIOUS_Project\no_table_v2"
    output_file = r"C:\Users\usert\Desktop\KHURIOUS_Project\no_table_v2"
    merge_markdown_files(input_directory, output_file)
```

```python
def split_md_file(input_file, output_prefix, num_parts=3):
    # MD 파일 읽기
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    
    # 총 줄 수 및 분할 크기 계산
    total_lines = len(lines)
    lines_per_part = total_lines // num_parts
    
    # 분할 및 저장
    for i in range(num_parts):
        start = i * lines_per_part
        # 마지막 파일에 남은 모든 줄 추가
        end = (i + 1) * lines_per_part if i < num_parts - 1 else total_lines
        
        part_lines = lines[start:end]
        output_file = f"{output_prefix}_part{i+1}.md"
        
        with open(output_file, 'w', encoding='utf-8') as out_f:
            out_f.writelines(part_lines)
        
        print(f"파일 저장 완료: {output_file}")

split_md_file('2024교육과정_eng_md.txt', 'output_file')
```
